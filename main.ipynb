1: Extract input and output cols 
    In our data, IQ & CGPA is INPUT COLS and Placement is Output cols as it is dependent on input cols 

```python
x = df.iloc[:,0:2]
y = df.iloc[:,-1]
```

3: Train Test Split     

```python
from sklearn.model_selection import train_test_split
```

```python
pip install scikit-learn
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1) # While running this we get 4 Values X of training set, X of test set, Y of Training set and Y of Test set
# SO WE WILL STORE THESE VALUES IN THEIR RESPECTIVE COLUMNS
```

```python
x_train
```

```python
y_train
```

#2: Scale the values

```python
from sklearn.preprocessing import StandardScaler # StandardScaler is a Class 
```

```python
scaler = StandardScaler()     # Creating object of StandardScaler class  
```

```python
x_train = scaler.fit_transform(x_train)   #fit transfrom will first Understand the data and than Transform it 
```

```python
x_train
```

```python
#Now same we will do for x_test
x_test = scaler.transform(x_test)
```

```python
x_test
```

#4: Train the model 
    As we have decided we are going to use Logistic Regression Classifier that is Model based ML Type that divides data in Decision Boundary 

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf = LogisticRegression() #creating Object of Logistic Regression
```

```python
clf.fit(x_train,y_train) #fit function inside Logistic Regression Class is used to Train the Data, 
#we will provide training data in Paramaters 
```

WOAAHHH Our Model is Trained !!!!

Now we'll come on 
# 5: Evaluating the model 
 we'll now get the accuracy of model on out test data 

```python
clf.predict(x_test)
```
