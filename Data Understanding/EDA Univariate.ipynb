# EDA (Univariate Analysis)

Univariate analysis is the simplest form of data analysis, focusing on examining one variable at a time. It describes and summarizes a single characteristic or attribute within a dataset without looking at causes or relationships between multiple variables

Variable = Column

a. Countplot

```python
#sns.countplot(df['Sex'])
sns.countplot(fd['internet_access'])
```

```python
fd['internet_access'].value_counts()
```

```python
sns.countplot(fd['study_method'])
```

```python
fd['study_method'].value_counts()
```

```python
fd['final_grade'].value_counts().plot(kind='bar')
```

b. Pie Chart
